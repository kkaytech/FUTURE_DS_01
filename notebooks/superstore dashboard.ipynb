```python
import pandas as pd
df = pd.read_csv("FUTURE_DS_01/data/superstore.csv", encoding='latin1')
df.head()
```

```python
df.info()
df.head()
```

```python
# Convert dates 
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])

# Remove duplicates 
df = df.drop_duplicates()

df.info()
```

```python
# Total Sales
total_sales = df['Sales'].sum()

# Total Profit
total_profit = df['Profit'].sum()

# Profit Margin
profit_margin = total_profit / total_sales

print("Total Sales:", round(total_sales, 2))
print("Total Profit:", round(total_profit, 2))
print("Profit Margin:", round(profit_margin, 4))
```

```python
top_products = df.groupby('Product Name')['Sales'].sum() \
                 .sort_values(ascending=False).head(10)

top_products
```

```python
top_region = df.groupby('Region')['Profit'].sum() \
               .sort_values(ascending=False)

top_region
```

```python
top_category = df.groupby('Category')['Sales'].sum() \
                 .sort_values(ascending=False)

top_category
```

```python
df['Month-Year'] = df['Order Date'].dt.to_period('M')

monthly_sales = df.groupby('Month-Year')['Sales'].sum()

monthly_sales
```

```python
segment_profit = df.groupby('Segment')['Profit'].sum() \
                   .sort_values(ascending=False)

segment_profit
```

```python
# Check discount impact
discount_profit = df.groupby('Discount')['Profit'].mean().sort_values()

discount_profit
```

## **Key Business Insights**
### 1. Revenue Drivers
The Technology category generates the highest revenue (~836K), making it the primary growth driver.
Furniture and Office Supplies follow closely, but neither surpass Technology in contribution.

**Insight**:
Technology products should remain a core focus for expansion and marketing investment.

### 2. Regional Performance
The West region delivers the highest profit (~108K), significantly outperforming others.
Central region shows the lowest profitability.

**Insight**:
- West is the strongest market
- Central may need cost optimization or pricing review

### 3. Product-Level Performance
High-value products like Canon imageCLASS Copier and binding machines dominate sales.

**Insight**:

- A small number of products drive a large portion of revenue
- This indicates a high dependency on top-performing SKUs
  
### 4. Segment Analysis
Consumer segment contributes the highest profit (~134K)
Corporate and Home Office follow

**Insight**:

- Consumer market is the most profitable and stable segment
- Corporate could be optimized further with targeted pricing strategies

### 5. Profitability Overview
Overall profit margin ≈ 12.47%

**Insight**:

- Healthy margin, but not very high
- Suggests possible discount-heavy strategy or high costs

### Critical Insight:
*Profitability declines sharply as discount rates increase. Discounts above 20% are associated with negative average profit, suggesting that aggressive discounting is hurting margins rather than driving profitable growth. Although the business remains profitable overall, deeper discount levels are associated with weaker or negative profitability, indicating that aggressive discounting is eroding margins in specific transactions or product groups.*



## **Recommendations:**
### 1. Optimize Discount Strategy
- Reduce excessive discounting, especially in low-margin categories
- Implement profit-based discount limits
  
### 2. Focus on High-Performing Regions
- Invest more in the West region
- Analyze why Central underperforms and fix pricing/logistics
  
### 3. Expand Technology Category
- Increase inventory and marketing for Technology products
- Bundle high-performing products to increase basket size
  
### 4. Improve Furniture Profitability
- Furniture has high sales but relatively lower profit
- Review: pricing, shipping costs, supplier margins
  
### 5. Strengthen Corporate Segment
- Introduce targeted offers or contracts
- Improve long-term B2B relationships
